```python
import numpy as np
import matplotlib.pyplot as plt
 
# set width of bar
barWidth = 0.25
 
# set height of bar
bars1 = [0.5, 0.3]
bars2 = [0.2, 0.6]
bars3 = [0.3, 0.1]
 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
 
# Make the plot
plt.bar(r1, bars1, color='blue', width=barWidth, edgecolor='white', label='Class A')
plt.bar(r2, bars2, color='darkblue', width=barWidth, edgecolor='white', label='Class B')
plt.bar(r3, bars3, color='lightblue', width=barWidth, edgecolor='white', label='Class C')

plt.ylabel('Proportion of class')
# Add xticks on the middle of the group bars
#plt.xlabel('group', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(bars1))], ['Training', 'Testing'])
 
# Create legend & Show graphic
plt.legend()
plt.show()
```
